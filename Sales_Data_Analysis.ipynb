FIXING OUTLIERS


```python
def Iqr(x):
    q1=x.quantile(.25)
    q3=x.quantile(.75)
    IQR=q3-q1
    Lf=q1-(1.5*IQR)
    Uf=q3+(1.5*IQR)
    print("Lf-->",Lf,"Uf-->",Uf)
```

```python
Iqr(df['Total profit'])
```

```python
df['Total profit']=np.where(df['Total profit']>1284028.79375,1284028.79375,df['Total profit'])
plt.boxplot(df['Total profit'])
plt.show()
```
